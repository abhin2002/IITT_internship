```python
import cv2
import numpy as npy
import os

background_list = [12, 13, 14, 16, 17, 18, 19, 20]
background_list2 = [4, 5, 6]

for video in background_list2:
    # Define the paths to the directories containing the images
    foreground_dir = '/mnt/DATA/dronetracking/final_composition/foreground4/video03/foreground_image'
    background_dir = f'/mnt/DATA/dronetracking/Anti-UAV-Tracking-V0/video0{video}'
    mask_dir = '/mnt/DATA/dronetracking/final_composition/foreground4/video03/original_sized_mask'
    composition_path = f'/mnt/DATA/dronetracking/Abhin/final_composition/compostion_result/foreground_video3_compo/comp1_0{video}'

    os.makedirs(composition_path, exist_ok=True)

    # List all files in the directories
    foreground_files = sorted(os.listdir(foreground_dir))
    background_files = sorted(os.listdir(background_dir))
    background_files = [file for file in background_files if file.endswith('.jpg')]
    mask_files = sorted(os.listdir(mask_dir))

    min_files = min(len(background_files), len(foreground_files), len(mask_files))

    # Process each image pair until you run out of background images
    for i in range(min_files):
        foreground_file = foreground_files[i]
        background_file = background_files[i]
        mask_file = mask_files[i]

        # Load the images
        foreground_path = os.path.join(foreground_dir, foreground_file)
        background_path = os.path.join(background_dir, background_file)
        mask_path = os.path.join(mask_dir, mask_file)

        foreground = cv2.imread(foreground_path)
        background = cv2.imread(background_path)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)  # Load mask as grayscale

        # Resize the foreground and mask to match the background size if necessary
        foreground_resized = cv2.resize(foreground, (background.shape[1], background.shape[0]))
        mask_resized = cv2.resize(mask, (background.shape[1], background.shape[0]))

        # Convert segmentation mask to binary
        _, mask_binary = cv2.threshold(mask_resized, 1, 255, cv2.THRESH_BINARY)

        # Invert the mask
        mask_inv = cv2.bitwise_not(mask_binary)

        # Apply the inverted mask to the background
        background_bg = cv2.bitwise_and(background, background, mask=mask_inv)

        # Add the images
        combined_image = cv2.add(foreground_resized, background_bg)

        # Save the combined image
        output_path = os.path.join(composition_path, f'combined_{(i+1):05d}.png')  # Change to your output directory
        cv2.imwrite(output_path, combined_image)

        print(f"Processed {foreground_file}, {background_file}, {mask_file}")
    print(f'background {video} completed')

```
