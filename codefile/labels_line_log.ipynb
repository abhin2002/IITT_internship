```python
import os

def log_file_names(folder_path):
    # Paths for the output log files
    single_line_log = 'single_line_files.txt'
    multi_line_log = 'multi_line_files.txt'

    # Lists to store filenames
    single_line_files = []
    multi_line_files = []

    # Iterate through all files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file is a text file
        if filename.endswith('.txt'):
            file_path = os.path.join(folder_path, filename)
            
            # Open and read the file
            with open(file_path, 'r') as file:
                lines = file.readlines()
                if len(lines) == 1:
                    single_line_files.append(filename)
                else:
                    multi_line_files.append(filename)

    # Write the filenames to the respective log files
    with open(single_line_log, 'w') as file:
        for name in single_line_files:
            file.write(name + '\n')

    with open(multi_line_log, 'w') as file:
        for name in multi_line_files:
            file.write(name + '\n')

# Example usage
folder_path = '/mnt/DATA/dronetracking/Abhin/segement&detect/detection_DUT/train/labels'
log_file_names(folder_path)
```

```python
import os
import shutil

def log_and_copy_files(text_folder, image_folder, single_line_text_output_folder, single_line_image_output_folder):
    # Path for the output log file
    single_line_log = 'single_line_files.txt'
    
    # Create output folders if they don't exist
    os.makedirs(single_line_text_output_folder, exist_ok=True)
    os.makedirs(single_line_image_output_folder, exist_ok=True)

    # List to store filenames
    single_line_files = []

    # Iterate through all files in the text folder
    for filename in os.listdir(text_folder):
        # Check if the file is a text file
        if filename.endswith('.txt'):
            file_path = os.path.join(text_folder, filename)
            
            # Open and read the file
            with open(file_path, 'r') as file:
                lines = file.readlines()
                if len(lines) == 1:
                    single_line_files.append(filename)
                    # Copy the text file to the single_line_text_output_folder
                    shutil.copy(file_path, single_line_text_output_folder)
                    
                    # Copy the corresponding image file
                    image_filename = filename.replace('.txt', '.jpg')  # Assuming the image files are in .jpg format
                    image_path = os.path.join(image_folder, image_filename)
                    if os.path.exists(image_path):
                        shutil.copy(image_path, single_line_image_output_folder)

    # Write the filenames to the log file
    with open(single_line_log, 'w') as file:
        for name in single_line_files:
            file.write(name + '\n')

# Example usage
text_folder = '/mnt/DATA/dronetracking/Abhin/segement&detect/detection_DUT/val/mask_polygon/first train data/labels'
image_folder = '/mnt/DATA/dronetracking/Abhin/segement&detect/detection_DUT/val/mask_polygon/first train data/images'
single_line_text_output_folder = '/mnt/DATA/dronetracking/Abhin/segement&detect/detection_DUT/val/labels'
single_line_image_output_folder = '/mnt/DATA/dronetracking/Abhin/segement&detect/detection_DUT/val/images'

log_and_copy_files(text_folder, image_folder, single_line_text_output_folder, single_line_image_output_folder)
```
