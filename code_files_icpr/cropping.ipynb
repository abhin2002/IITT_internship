```python
import matplotlib.image as mpimg
from matplotlib.image import imsave
import os
import matplotlib.image as mpimg
from skimage.io import imsave
```

```python

# Output folder path for the cropped images
output_folder_path = "/mnt/DATA/dronetracking/Abhin/segement&detect/drone_detection_icpr_new/test/cropped"

# Directory containing the images
folder_path = '/mnt/DATA/dronetracking/Abhin/segement&detect/drone_detection_icpr_new/test/images'

# Directory containing the bounding box files
bbox_folder_path = '/mnt/DATA/dronetracking/Abhin/segement&detect/drone_detection_icpr_new/test/labels'

# Get a list of all image files in the directory
file_list = sorted([f for f in os.listdir(folder_path) if f.endswith('.png')])

# Loop through each image file and process the image with the corresponding bounding box
for idx, file_name in enumerate(file_list, start=1):
    image_path = os.path.join(folder_path, file_name)
    image = mpimg.imread(image_path)
    
    # Corresponding bounding box file path
    bbox_file_path = os.path.join(bbox_folder_path, f'{os.path.splitext(file_name)[0]}.txt')
    
    # Read the bounding box from the text file
    with open(bbox_file_path, 'r') as bbox_file:
        bbox = bbox_file.readline().strip()
    
    # Parse the bounding box coordinates
    dr_id, x, y, w, h = map(float, bbox.split())
    
    # Calculate the extended cropping coordinates
    y_min = int(y - (h / 4))
    y_max = int(y + h + (h / 4))
    x_min = int(x - (w / 4))
    x_max = int(x + w + (w / 4))
    
    # Ensure the coordinates are within the image bounds
    y_min = max(0, y_min)
    y_max = min(image.shape[0], y_max)
    x_min = max(0, x_min)
    x_max = min(image.shape[1], x_max)
    
    # Crop the image
    image_c = image[y_min:y_max, x_min:x_max]
    
    # Save the cropped image
    output_file_name = f'{idx:05d}_cropped.png'
    output_file_path = os.path.join(output_folder_path, output_file_name)
    imsave(output_file_path, image_c)
    print(image_c)
```
