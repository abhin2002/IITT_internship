```python
orginal_image = f"/mnt/DATA/dronetracking/Anti-UAV-Tracking-V0/{video_name}/00001.jpg"
output_folder_path = f"/mnt/DATA/dronetracking/final_composition/foreground/{video_name}/original_sized_mask"
seg_folder_path = f'/mnt/DATA/dronetracking/final_composition/foreground/{video_name}/cropped_mask'
bbox_file_path = f'/mnt/DATA/dronetracking/Anti-UAV-Tracking-V0GT/{video_name}_gt.txt'

original_height, original_width, dimension = mpimg.imread(orginal_image).shape
print(original_height, original_width)
```

```python
with open(bbox_file_path, 'r') as bbox_file:
    bboxes = bbox_file.readlines()

seg_file_list = sorted([f for f in os.listdir(seg_folder_path) if f.endswith('.png')])

# plotting_interval = 500
count = 0
# Loop through each file and process the image with the bounding box
for idx, (file_name, bbox) in enumerate(zip(seg_file_list, bboxes), start=1):
    # For purpose of checking the correctness stopping the creation 
    
    count += 1
    crop_seg_path = os.path.join(seg_folder_path, file_name)
    crop_seg = mpimg.imread(crop_seg_path)
    # # Display the image
    # plt.imshow(image, cmap='gray')
    # plt.axis('off')  # Hide axis ticks and labels
    # plt.show()
    
    
    # Parse the bounding box coordinates
    x, y, w, h = map(int, bbox.strip().split())
    
    y_min = int(y - (h / 4))
    y_max = int(y + h + (h / 4))
    x_min = int(x - (w / 4))
    x_max = int(x + w + (w / 4))
    
    image_o = np.zeros((original_height, original_width, dimension))
    
    if(image_o[y_min:y_max, x_min:x_max].shape != crop_seg.shape) :
        print(f"{count} : {image_o[y_min:y_max, x_min:x_max].shape}  {crop_seg.shape}")
    else:
        image_o[y_min:y_max, x_min:x_max] = crop_seg
    
    # Save the cropped image
    output_file_name = f'c_{idx:05d}.png'
    output_file_path = os.path.join(output_folder_path, output_file_name)
    imsave(output_file_path, image_o)
    
    # # Display the cropped image at the specified interval
    # if idx % plotting_interval == 0:
    #     fig, ax = plt.subplots()
    #     ax.imshow(image_o)
    #     plt.axis('off')
    #     plt.show()
```
